```python
from sympy import *
from sympy.abc import x,n
```

```python
# Evaluate the following integrals using a computer algebra system
```

```python
a = Integral(x*E**x)
a
```

```python
b = Integral(x**2*E**x)
b
```

```python
c = Integral(x**3*E**x)
c
```

```python
d = Integral(x**4*E**x)
d
```

```python
e = Integral(x**5*E**x)
e
```

```python
# Solutions to the above equations:
```

```python
Eq(a,a.doit())
```

```python
Eq(b,b.doit())
```

```python
Eq(c,c.doit())
```

```python
Eq(d,d.doit())
```

```python
Eq(e,e.doit())
```
