```python
from sympy import *
from sympy.abc import x,n
```

```python
# Evaluate the following integrals using a computer algebra system
```

```python
a = Integral(ln(x))
a
```

```python
b = Integral(ln(x)*x)
b
```

```python
c = Integral(ln(x)*x**2)
c
```

```python
d = Integral(ln(x)*x**3)
d
```

```python
e = Integral(ln(x)*x**7)
e
```

```python
# Solutions to the above equations:
```

```python
Eq(a,a.doit())
```

```python
Eq(b,b.doit())
```

```python
Eq(c,c.doit())
```

```python
Eq(d,d.doit())
```

```python
Eq(e,e.doit())
```
